```python
import csv
from openpyxl.workbook import Workbook


class Record:
    def __init__(self, dict_data: dict[str, str]):
        self.__dict_data = dict_data

    @property
    def year(self):
        return self.__dict_data["年份"]

    @property
    def values(self):
        return self.__dict_data.values()

    @property
    def keys(self):
        return self.__dict_data.keys()

    def __str__(self):
        return str(self.__dict_data)


class DataUtils:
    @staticmethod
    def load_records(csv_path: str):
        """
        加载CSV文件中的记录
        """
        with open(csv_path, encoding="utf-8-sig") as csvfile:
            reader = csv.DictReader(csvfile)
            records = []
            for row in reader:
                records.append(Record(row))
            return records

    @staticmethod
    def group_by_year(records: list[Record]) -> dict[str, list[Record]]:
        """
        根据年份将记录拆分为多个列表
        """
        by_year = {}
        for record in records:
            year = record.year
            if year in by_year:
                by_year[year].append(record)
            else:
                by_year[year] = [record]

        return by_year

    @staticmethod
    def append_to_workbook(workbook: Workbook, sheet_name: str, records: list[Record]):
        """
        将记录添加到工作簿中
        """

        if not records:
            return

        first = records[0]

        worksheet = workbook.create_sheet(sheet_name)
        worksheet.append(list(first.keys))

        for record in records:
            worksheet.append(list(record.values))


input_csv = "1686117568248-题目1数据.csv"
output_xlsx = "output.xlsx"

all_records = DataUtils.load_records(input_csv)
records_by_year = DataUtils.group_by_year(all_records)

wb = Workbook(True)

for year, records in records_by_year.items():
    DataUtils.append_to_workbook(wb, year, records)

wb.encoding = "utf-8-sig"
wb.save(output_xlsx)
wb.close()

print("转换完成。输出文件：", output_xlsx)
```
