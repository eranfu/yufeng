```python
from numpy.ma.testutils import assert_equal


def kth_largest(nums, k):
    """
    Find the kth largest number in nums.
    Use quickselect algorithm.
    :param nums: The numbers.
    :param k: Start from 1.
    :return: The kth largest number.
    """

    def swap(i, j):
        if i != j:
            nums[i], nums[j] = nums[j], nums[i]

    def partition(start, end):
        """
        Partition nums[start: end] into three parts, nums[start: pivot], nums[pivot] and nums[pivot + 1: end].

        1. nums[start: pivot] contains the numbers greater than pivot_value.

        2. nums[pivot] is the pivot_value.

        3. nums[pivot + 1: end] contains the numbers less than or equal to pivot_value.

        :return: the pivot
        """

        if start >= end:
            raise ValueError(f"start({start}) >= end({end})")

        if end - start == 1:
            return start

        pivot_value = nums[end - 1]
        store_index = start
        for i in range(start, end - 1):
            if nums[i] > pivot_value:
                swap(i, store_index)
                store_index += 1
        swap(end - 1, store_index)
        return store_index

    def quickselect(start, end):
        pivot = partition(start, end)
        if pivot == k - 1:
            return nums[pivot]
        elif pivot > k - 1:
            return quickselect(start, pivot)
        else:
            return quickselect(pivot + 1, end)

    if k < 1 or k > len(nums):
        return None

    return quickselect(0, len(nums))
```

```python
# 作业题目用例
assert_equal(kth_largest([3,2,1,5,6,4], 2), 5)
assert_equal(kth_largest([1], 1), 1)
assert_equal(kth_largest([3, 2], 1), 3)
assert_equal(kth_largest([3, 2], 2), 2)

# 边界情况
assert_equal(kth_largest([1,2,3], 0), None)
assert_equal(kth_largest([1,2,3], 4), None)
assert_equal(kth_largest([3, 2], 3), None)
assert_equal(kth_largest([], 1), None)
assert_equal(kth_largest([5, 4, 3, 2, 1], 1), 5)
assert_equal(kth_largest([5, 4, 3, 2, 1], 5), 1)

# 包含负数的情况
assert_equal(kth_largest([-1, -2, -3, -4, -5], 1), -1)
assert_equal(kth_largest([-1, -2, -3, -4, -5], 5), -5)
assert_equal(kth_largest([10, -5, 0, 15, -10], 1), 15)
assert_equal(kth_largest([10, -5, 0, 15, -10], 3), 0)

# 重复元素
assert_equal(kth_largest([1, 1, 2, 3, 3], 2), 3)
assert_equal(kth_largest([1, 1, 2, 3, 3], 3), 2)
assert_equal(kth_largest([1, 2, 3, 4, 5, 5], 2), 5)
assert_equal(kth_largest([1, 2, 3, 4, 5, 5], 3), 4)
assert_equal(kth_largest([1, 2, 3, 4, 5, 5], 1), 5)
assert_equal(kth_largest([1, 1, 1, 1, 1], 1), 1)
assert_equal(kth_largest([1, 1, 1, 1, 1], 3), 1)
assert_equal(kth_largest([2, 2, 1, 1, 3, 3], 1), 3)
assert_equal(kth_largest([2, 2, 1, 1, 3, 3], 2), 3)
assert_equal(kth_largest([2, 2, 1, 1, 3, 3], 3), 2)
assert_equal(kth_largest([5,5,5,5,5], 3), 5)

# 已排序数组（正序和倒序）
assert_equal(kth_largest([1,2,3,4,5], 3), 3)  # 正序
assert_equal(kth_largest([5,4,3,2,1], 3), 3)  # 倒序

# 分区边界测试
assert_equal(kth_largest([3,1,4,1,5], 3), 3)  # pivot正好是第k大

# 大数组
large_array = list(range(100, 0, -1))  # 100到1的倒序
assert_equal(kth_largest(large_array, 1), 100)
assert_equal(kth_largest(large_array, 50), 51)
assert_equal(kth_largest(large_array, 100), 1)

# 随机顺序
random_array = [7, 3, 9, 1, 8, 2, 6, 4, 5]
assert_equal(kth_largest(random_array, 1), 9)
assert_equal(kth_largest(random_array, 5), 5)
assert_equal(kth_largest(random_array, 9), 1)

print("所有测试用例通过！")
```
