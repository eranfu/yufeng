```python
import pandas as pd
import numpy as np

dates1 = pd.date_range(start='2026-01-01', end='2026-1-31', freq='D').to_series().sample(5).sort_values()
dates2 = pd.date_range(start='2025-04-01', end='2025-4-30', freq='D').to_series().sample(3).sort_values()

order_id1 = np.random.randint(10000, 49999 + 1, size=5)
order_id2 = np.random.randint(50000, 99999 + 1, size=3)

products = ['A', 'B', 'C', 'D', 'E']
product1 = np.random.choice(products, size=5)
product2 = np.random.choice(products, size=3)

money1 = np.random.randint(100, 1000 + 1, size=5)
money2 = np.random.randint(100, 1000 + 1, size=3)

df1 = pd.DataFrame({'date': dates1, 'order_id': order_id1, 'product': product1, 'money': money1})
df2 = pd.DataFrame({'date': dates2, 'order_id': order_id2, 'product': product2, 'money': money2})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```
